```python
def times3plus1(num:str):
    nxt = ''
    c = 0
    for a, b in reversed(list(zip('0'+num, num+'0'))):
        if a == b:
            nxt += '1' if c else '0'
            c = 0 if a == '0' else 1
        else:
            nxt += '0' if c else '1'
    if c:
        nxt += '1'
    nxt += '0'
    i = nxt.index('0')
    nxt = i*'0' + '1' + nxt[i+1:]
    return nxt[::-1]
    
```

```python
def findNxt(num:str):
    try:
        i = list(num)[::-1].index('0')
        return num[:-1-i] + '1' + '0' * i
    except:
        return '1' + '0' * len(num)
```

```python
from random import random
def gen(length:int):
    num = '1'
    for _ in range(length-2):
        num += '1' if random() >= 0.5 else '0'
    return num + '1'
```

```python
def calc(num:str):
    ct = 0
    while num != '1':
        ct += 1
        num = times3plus1(num)
        num = num.strip('0')
        #print(num, int(num, 2))
    return ct
```

```python
from collections import defaultdict
count = defaultdict(int)

num = "1" + "0" * 69
for _ in range(100):
    num = findNxt(num)
    count[calc(num)] += 1
print(count)
```
